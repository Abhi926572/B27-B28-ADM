```python
data = pd.read_csv("Diabetes Missing Data.csv")
print("\nOriginal DataFrame with Missing values :")
print(data.info())
print(data.isnull().sum())
if 'BMI' in data.columns:
  data['BMI']=data['BMI'].fillna(data['BMI'].mean())
print(data.isnull().sum())
if 'Skin_Fold' in data.columns:

   data['Skin_Fold']=data['Skin_Fold'].fillna(data['Skin_Fold'].median())
if 'Glucose' in data.columns:

     data['Glucose']=data['Glucose'].fillna(data['Glucose'].median())

if 'Serum_Insulin' in data.columns:

    data['Serum_Insulin']=data['Serum_Insulin'].fillna(data['Serum_Insulin'].median())
if 'Diastolic_BP' in data.columns:
  data['Diastolic_BP']=data['Diastolic_BP'].fillna(data['Diastolic_BP'].median())

print(data.isnull().sum())

```
